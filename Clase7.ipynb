```python
def margen(sales,costs) -> float:
  """Calcula el margen entre ventas y costos de un
  producto retornando un valor tipo float"""

  return (sales - costs) / sales
```

```python
def is_premium(product:int) -> bool:
  """Revisa si un producto es de calidad premium devolviendo
  un True si el producto supera la nota de 3"""
  return True if product >= 3 else False
```

```python
is_premium(5)
margen(1000,80)
```

Crear una funcin recursiva que cuente regresivamente desde un nmero dado hasta 0.

```python
def count_down(number):
  if number == 0:
    return number
  print(number)
  return count_down(number - 1)
```

```python
count_down(10)
```
