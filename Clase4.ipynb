```python
campayne_list = {
    'campain_1' : {
        'clicks': 120,
        'prints': 8000,
        'rate': 25,
        'budget' : 300

    },
    'campain_2' : {
        'clicks': 90,
        'prints': 2000,
        'rate': 10,
        'budget' : 150
    }
}

results = {}

for x in campayne_list:
    ctr = (campayne_list[x]['clicks'] / campayne_list[x]['prints']) * 100
    tasa = (campayne_list[x]['rate'] / campayne_list[x]['clicks']) * 100
    if ctr <1:
      clasification = 'bad'
    elif ctr > 1 and ctr < 3:
      clasification = 'regular'
    else:
      clasification = 'good'
    results[x] ={'ctr': round(ctr,2),'tasa':round(tasa,2),'clasification':clasification}

results
```
